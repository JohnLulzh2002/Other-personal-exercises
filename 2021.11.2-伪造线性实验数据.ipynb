```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(14,4,-1)
print(x)
```

```python
l=np.array([26.975,26.712,26.572,26.402,26.232,26.122,25.942,25.752,25.536,25.358])
r=np.array([20.602,20.732,20.862,20.982,21.128,21.308,21.476,21.638,21.818,21.926])
```

```python
plt.plot(x,l)
plt.show()
plt.plot(x,r)
plt.show()
```

```python
lk,lb=np.polyfit(x,l,1)
rk,rb=np.polyfit(x,r,1)
print(lk,lb)
```

```python
plt.scatter(x,l)
lHat=lk*x+lb
plt.plot(x,lHat)
plt.show()
plt.scatter(x,r)
rHat=rk*x+rb
plt.plot(x,rHat)
plt.show()
```

```python
lError=l-lHat
rError=r-rHat
lMu=np.mean(lError)
rMu=np.mean(rError)
lSigma=np.std(lError,ddof=1)
rSigma=np.std(rError,ddof=1)
lErrorHat=lMu+np.random.randn(x.size)*lSigma
rErrorHat=rMu+np.random.randn(x.size)*rSigma
plt.scatter(x,lError)
plt.scatter(x,lErrorHat)
plt.axhline(0)
plt.show()
```

```python
print(lHat+lErrorHat)
print(rHat+rErrorHat)
```
